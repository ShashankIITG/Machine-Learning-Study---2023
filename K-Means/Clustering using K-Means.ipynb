```python
def distance(p1, p2):
    return np.sqrt(np.sum((p1-p2)**2))
```

```python
class Kmeans:
    def __init__(self, X, k):
        datapoints = X.shape[0]
        dimension = X.shape[1]
        centers = np.random.uniform(low = np.min(X), high = np.max(X), size = (k, dimension))
        clusters = {}
        label = []

    def assignPoints2Cluster(clusters):

        for i in range(k):
            cluster = {
                'center': centers[i],
                'points': []
            }
            clusters[i] = cluster

        for ix in range(datapoints):
            dist = []
            datapoint = X[ix]

            for center in range(k): dist.append(distance(center, datapoint))
            ic = np.argmin(dist)

            label.append(ic)
            clusters[ic]['points'].append(datapoint)
        
    def updateCluster(k, clusters):
        for kx in range(k):
            pts = np.array(clusters[kx]['points'])

            if pts.shape[0]>0: # if a cluster have non-zero points
                new_u = pts.mean(axis=0)
                clusters[kx]['center'] = new_u
                clusters[kx]['points'] = []
            
    def plotClusters(clusters):
        plt.grid(True)
        
        for kx in range(5):
            pts = np.array(clusters[kx]['points'])
            plt.scatter(pts[:,0], pts[:,1])

            ctr = np.array(clusters[kx]['center'])
            plt.scatter(ctr[0], ctr[1], marker = "*")
        plt.show
```
