# PCA - Principal Component Analysis

- idea -> retain most of the variance upon dimensionality reduction

- higher variance should be explained

```python
import pandas as pd
```

```python
dataset = pd.read_csv('./mnist dataset/mnist_train.csv')
```

```python
cut = int(dataset.shape[0]*.8)
train, test = dataset[:cut], dataset[cut:]
```

```python
X_train, Y_train = train.iloc[:,1:train.shape[1]], train['label']
X_test, Y_test = test.iloc[:,1:test.shape[1]], test['label']
```

```python
X_train
```

```python
Y_train
```

```python
print(X_train.shape, Y_train.shape)
print(type(X_train), type(Y_train))
```

```python
X_train = X_train.values
Y_train = Y_train.values

X_test = X_test.values
Y_test = Y_test.values
```

```python
## X_train = X_train.reshape((X_train.shape[0], 28, 28))
## X_test = X_test.reshape((X_test.shape[0], 28, 28))
```

```python
X_train.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(X_train[90].reshape((28, 28)), cmap = 'gray')
plt.show()
```

```python
## step 1 - Pre-Processing (Normalization)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
Xtrain = sc.fit_transform(X_train)
```

```python
Xtrain.shape
```

```python
plt.imshow(Xtrain[90].reshape((28,28)), cmap="gray")
plt.show()
```

# Sklearn PCA - Prebuild model

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
#X with 784 features to Z with 3 features
Z_pca = pca.fit_transform(Xtrain)
```

```python
Z_pca.shape
```

```python
pca.explained_variance_
```

###### Original data with 784 feature (pixel) have 100% of information/variance and upon dimensionality reduction to 3 features the information is reduced to 40+29+26 = 95% i.e., 95 variance

# Custom PCA

```python
import numpy as np
```

```python
# step 2 - compute covariance
```

```python
X_train.shape
```

```python
X_train_mean = X_train.mean(axis=0)
```

```python
X_train_mean.shape
```

```python
print(X_train_mean[:50])
```

```python
# some means are not zero
```

```python
X_diff = X_train - X_train_mean
```

```python
N = X_train.shape[1]
```

```python
cov = (1/N-1)*np.dot(X_diff.T, X_diff)
```

```python
cov.shape
```

```python
## step 3 - compute eigen vector using SVD
```

```python
from numpy.linalg import svd
```

```python
svd?
```

```python
U, S, V = svd(cov)
```

```python
U.shape
```

```python
Ured = U[:,:3]
```

```python
Ured.shape
```

```python
# step 4 - projection of dataset on 3 new axis (components)
```

```python
Z = np.dot(Xtrain, Ured)
```

```python
Z.shape
```

```python
Z
```

```python
Z_pca
```
