```python
X = pd.read_csv("../Dataset/Hardwork/Training Data/Linear_X_Train.csv")
Y = pd.read_csv("../Dataset/Hardwork/Training Data/Linear_Y_Train.csv")
```

```python
plt.style.use('seaborn')
plt.scatter(X, Y)
plt.xlabel("Hardwork")
plt.ylabel("Performance or Marks")
plt.title("Hardwork pays off")
plt.show()
```

```python
## convert dataframe to numpy array
X = X.values
Y = Y.values
```

```python
## normalization
u = X.mean()
std = X.std()
X = (X-u)/std
```

```python
def hypothesis(x, theta):
    y_ = theta[0] + theta[1]*x
    return y_
```

```python
## J(theta)
def error(X, Y, theta):
    m = X.shape[0]
    total_error = 0.0
    for i in range(m):
        x = X[i]
        y_ = hypothesis(x, theta)
        y = Y[i]
        total_error += (y_ - y)**2
        
    return total_error/m
```

```python
# delta of J(theta)
def gradient(X, Y, theta):
    m = X.shape[0]
    grad = np.zeros((2,))
    for i in range(m):
        x = X[i]
        y_ = hypothesis(x, theta)
        y = Y[i]
        grad[0] += (y_ - y)
        grad[1] += (y_ - y)*x
        
    return grad/m
```

```python
def gradientDescent(X, Y, learning_rate = 0.1):
    theta = np.zeros((2,))
    max_steps = 100
    error_list = []
    theta_list = []
    for i in range(max_steps):
        grad = gradient(X, Y, theta)
        # theta[0] = theta[0] - learning_rate * grad[0]
        # theta[1] = theta[1] - learning_rate * grad[1]
        
        theta -= learning_rate * grad
        theta_list.append((theta[0], theta[1]))
        
        e = error(X, Y, theta)
        error_list.append(e)
    
    return theta, theta_list, error_list
```

```python
theta, theta_list, error_list = gradientDescent(X, Y)
```

```python
plt.plot(error_list)
plt.title("Reduction in error over iterations")
plt.xlabel("number of iterations")
plt.ylabel("J() or mean square error")
plt.show()
```

```python
# T0, T1 = np.meshgrid(theta_list[:,0], theta_list[:,1])
T0 = np.arange(-40, 40, 1)
T1 = np.arange(40, 120, 1)
T0, T1 = np.meshgrid(T0, T1)

J = np.zeros(T0.shape)

for i in range(J.shape[0]):
    for j in range(J.shape[1]):
        Y_ = T1[i, j]*X + T0[i, j]
        J[i, j] = np.sum((Y_ - Y)**2)/Y.shape[0]
```

# Compute Score
Score: R2 (R-Squared) or Coefficient of Determination

```python
def r2_score(Y, Y_):
    num = np.sum((Y_ - Y)**2)
    denomi = np.sum((Y.mean() - Y)**2)
    
    score = 1 - num/denomi
    
    return score*100
```

```python
r2_score(Y, Y_)
```
