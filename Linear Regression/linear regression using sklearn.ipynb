```python
from sklearn import datasets
```

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
X, Y = datasets.make_regression(n_samples=1000, n_features = 4, n_informative = 2, noise = 10, random_state = 1)
```

```python
print(X.shape, Y.shape)
```

```python
plt.subplot(1, 4, 1)
plt.scatter(X[:,0], Y)

plt.subplot(1, 4, 2)
plt.scatter(X[:,1], Y)

plt.subplot(1, 4, 3)
plt.scatter(X[:,2], Y)

plt.subplot(1, 4, 4)
plt.scatter(X[:,3], Y)

plt.show()
```

```python
%matplotlib notebook
from mpl_toolkits import mplot3d

fig = plt.figure(figsize = (10, 7))
axes = plt.axes(projection = '3d')

axes.scatter3D(X[:,0], X[:,3], Y, color = 'green')
plt.title("3D scatter plot")
plt.show()
```

# Implementing Linear Regression Using Scikit-Learn

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict((X[0], X[1]))
```

```python
Y[0:2]
```

```python
model.score(X, Y)
```
