```python
df_dataset = pd.read_csv("../Datasets/car sales.csv")

# train test split
df_train = df_dataset[:-6]
df_test = df_dataset[-6:]

# pandas data frame into numpy array
train = df_train.values
test = df_test.values

# convert time data into integer for linear regression

def time2int(npdata):
    # split date time
    datetime = np.char.split(npdata[:,0].astype(str), sep = '-')
    datetime = np.vstack(datetime)
    
    # year = datetime[:, 0].astype(int)
    # month = datetime[:, 1].astype(int)
    datetime = datetime.astype(int)
    
    return datetime

X = time2int(train)
X = np.array(X, dtype='int')

Y = train[:, 1]


#Normalization
X = X - X.mean(axis=0) / X.std(axis=0)

## adding X0 for theta0 to the X array
X0 = np.ones((X.shape[0], 1))
X = np.hstack((X0, X))
```

```python
def hypothesis(X, theta):
    return np.dot(X, theta)
```

```python
def error(X, Y, theta):
    e = 0.0
    m = X.shape[0]
    Y_ = hypothesis(X, theta)
    e = np.sum((Y - Y_)**2)
    
    return e/m
```

```python
def gradient(X, Y, theta):
    m = X.shape[0]
    Y_ = hypothesis(X, theta)
    grad = np.dot(X.T, (Y_ - Y))
    
    return grad/m
```

```python
def gradientDescent(X, Y, learning_rate = 0.1, max_epochs = 300):
    m = X.shape[0]
    n = X.shape[1]
    
    theta = np.zeros((n,))
    theta_list = []
    error_list = []
    
    for i in range(max_epochs):
        grad = gradient(X, Y, theta)
        #print(grad.shape)
        theta = theta - (learning_rate*grad)
        theta_list.append(theta)
        
        e = error(X, Y, theta)
        error_list.append(e)
        
    return theta_list, error_list
```

```python
theta_list, error_list = gradientDescent(X, Y, max_epochs = 1000)
```

```python
Y_ = hypothesis(X, theta_list[999])
```

```python
def r2_score(Y_, Y):
    num = np.sum((Y_ - Y)**2)
    denomi = np.sum((Y.mean() - Y)**2)
    
    score = 1 - num/denomi
    
    return score*100
```
