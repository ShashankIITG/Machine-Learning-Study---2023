```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv("movie_metadata.csv")
```

```python
data_frame
```

```python
data_frame.columns
```

```python
data_frame.get('movie_title')
```

```python
titles = list(data_frame.get('movie_title'))
```

```python
titles
```

```python
freq_titles = {}

for title in titles:
    length = len(title)
    
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
```

```python
plt.scatter(X,Y)
plt.xlabel("length of movie titles")
plt.ylabel("Number of movies with titles length")
plt.title("Movie Data Visualization Plot")
plt.show()
```
