```python
def convolution(img, kernel):
    H, W = img.shape
    K = kernel.shape[0]
    
    new_img = np.zeros((H-K+1, W-K+1))
    
    for i in range(H-3):
        for j in range(W-3):
            new_img[i, j] = np.sum(kernel * img[i:i+3, j:j+3])
            if new_img[i, j] > 255:
                new_img[i, j] = 255
            elif new_img[i, j] < 0:
                new_img[i, j] = 0
    return new_img
```

```python
# image -> [convolution (filter/kernel) -> Relu (activation function) -> pooling] * repeat #n of times. 

def pooling(X, mode="max"):
    H, W = X.shape
    # filter size
    f = 2
    stride = 2
    
    #pooling output matrix height and weight
    Ho = int((H-f)/stride) + 1
    Wo = int((W-f)/stride) + 1
    output = np.zeros((Ho, Wo))
    
    for r in range(Ho):
        for c in range(Wo):
            X_slice = X[r*stride : r*stride+f+1, c*stride: c*stride+f+1]
            if mode == "max":
                output[r, c] = np.max(X_slice)
            elif mode == "mean":
                output[r, c] = np.mean(X_slice)
            else:
                output[r, c] = np.var(X_slice)
            
    return output
```

```python
# input to the pooling i.e., output from activation function (e.g., ReLU)
X = np.array([[1, 0, 2, 3], 
             [4, 6, 6, 8],
             [3, 1, 1, 0],
             [1, 2, 2, 4]])

pooling(X, "max")
```

```python
print(pooling(X, "max"))
print(pooling(X, "mean")) 
print(pooling(X, "var"))
```
