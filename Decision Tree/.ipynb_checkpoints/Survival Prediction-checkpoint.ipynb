### entropy and information gain

```python
def entropy(Y):
    counts = np.unique(Y, return_counts=True)
    ent = 0.0
    
    for freq in counts[1]:
        
        total = Y.shape[0]
        prob = freq / total
        
        ent += -1.0 * prob * np.log2(prob)
        
    return ent
```

```python
entropy(np.array([0, 0, 1, 1]))
```

```python
def divide_data(data, attribute, threshold):
    left = pd.DataFrame([], columns = data.columns)
    right = pd.DataFrame([], columns = data.columns)
    
    for ix in range(data.shape[0]):
        if data[attribute].loc[ix] < threshold:
            left = left.append(data.loc[ix])
        else:
            right = right.append(data.loc[ix])
    
    return left, right
```

```python
divide_data(train_df, 'Age', train_df['Age'].mean())
```

```python
def info_gain(data, data_entropy, attribute, threshold):
    
    ## IG(S) = H(S) - sum(Sv/S * H(Sv))
    left, right = divide_data(data, attribute, threshold)
    
    if left.shape[0] == 0 or right.shape[0] == 0:
        return -100000
    
    left_ratio = left.shape[0]/data.shape[0]
    right_ratio = right.shape[0]/data.shape[0]
    
    information_gain = data_entropy - (left_ratio * entropy(left.Survived) + right_ratio * entropy(right.Survived))
    
    return information_gain
```

```python
ent = entropy(train_df.Survived)

for fx in X.columns:
    information_gain = info_gain(train_df, ent, fx, train_df[fx].mean())
    print(fx, ":", information_gain)
```

### decision tree implementation

```python
info_gain
```

```python
class DecisionTree:
    
    def __init__(self, depth = 0, max_depth = 5):
        self.left = None
        self.right = None
        self.attribute = None
        self.threshold = None
        self.max_depth = max_depth
        self.depth = depth
        self.target = None
        
    def train(self, data):
        # stop condition
        if self.depth >= self.max_depth:
            if data.Survived.mean() > 0.5:
                self.target = 1.0
            else:
                self.target = 0.0
            return
            
        ent = entropy(data.Survived)
        information_gains = []
        
        attributes = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']

        ## set of features and remove the label
        # all_features = data.columns.values
        # remove_index = np.where( all_features == 'Survived')[0][0]
        # attributes = np.delete(all_features, remove_index)
        
        for fx in attributes:
            ig = info_gain(data, ent, fx, data[fx].mean())
            information_gains.append(ig)
            
        self.attribute = attributes[np.argmax(information_gains)]
        self.threshold = data[self.attribute].mean()
        print("Making tree node:", self.attribute)
        
        data_left, data_right = divide_data(data, self.attribute, self.threshold)
        
        # reset index of splitted data
        data_left = data_left.reset_index(drop=True)
        data_right = data_right.reset_index(drop=True)
        
        # checking leaf node condition - no split of data
        if data_left.shape[0] == 0 or data_right.shape[0] == 0:
            if data.Survived.mean() >= 0.5:
                self.target = 1.0
            else:
                self.target = 0.0
            return
        
        # recurssion
        self.left = DecisionTree(depth = self.depth + 1, max_depth = self.max_depth)
        self.left.train(data_left)
        
        self.right = DecisionTree(depth = self.depth + 1, max_depth = self.max_depth)
        self.right.train(data_right)
        
        # making decision/target at every node
        if data.Survived.mean() >= 0.5:
            self.target = 1.0
        else:
            self.target = 0.0
        return
    
    def predict(self, test):
        print(self.attribute, self.threshold)
        
        if test[self.attribute] > self.threshold:
            print(self.attribute, "right", self.right)
            if self.right.attribute is None:
                ## print(self.target, "right")
                return self.target
            return self.right.predict(test)
        else:
            print(self.attribute, "left", self.right)
            if self.left.attribute is None:
                ## print(self.target, "left")
                return self.target
            return self.left.predict(test)
```

```python
def create_tree_graph(tree, graph):
    if tree.left is not None:
        edge = pydotplus.Edge(str(tree.attribute), str(tree.left.attribute))
        graph.add_edge(edge)
        create_tree_graph(tree.left, graph)

    if tree.right is not None:
        edge = pydotplus.Edge(str(tree.attribute), str(tree.right.attribute))
        graph.add_edge(edge)
        create_tree_graph(tree.right, graph)
```

```python
def draw_tree(tree):
    graph = pydotplus.Dot(graph_type='graph')
    create_tree_graph(tree, graph)
    return graph
```
