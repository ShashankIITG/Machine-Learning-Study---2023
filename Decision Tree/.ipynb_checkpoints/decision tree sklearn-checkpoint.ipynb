```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree = DecisionTreeClassifier(criterion="entropy", max_depth=5)
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('titanic dataset/train.csv')
```

```python
col = ['Survived', 'Sex', 'Age', 'Pclass', 'SibSp', 'Parch', 'Fare']
```

```python
split = int(0.7 * data.shape[0])
train = data[:split]
test = data[split:]
```

```python
train = train[col]
test = test[col]
```

```python
train = train.fillna(train['Age'].mean())
test = test.fillna(train['Age'].mean())
```

```python
print(train.info(), test.info())
```

```python
out_col = ['Survived']
```

```python
int_col = ['Sex', 'Age', 'Pclass', 'Parch', 'SibSp', 'Fare']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

train['Sex'] = le.fit_transform(train['Sex'])
test['Sex'] = le.fit_transform(test['Sex'])
```

```python
decision_tree.fit(train[int_col], train[out_col])
```

```python
outputs = decision_tree.predict(test[int_col])
```

```python
decision_tree.score(test[int_col], test[out_col])
```
