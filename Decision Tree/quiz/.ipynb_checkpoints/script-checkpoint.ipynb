```python
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
```

```python
import numpy as np
```

```python
def entropy(Y):
    Y_unique = np.unique(Y)
    total = Y.shape[0]
    ent = 0.0
    
    for y in Y_unique:
        count = np.sum(Y == y)
        prob = count/total
        ent += -prob * np.log2(prob)
        
    return ent
```
