```python
import numpy as np
```

```python
def entropy(Y):
    Y_unique = np.unique(Y)
    total = Y.shape[0]
    ent = 0.0
    
    for y in Y_unique:
        count = np.sum(Y == y)
        prob = count/total
        ent += -prob * np.log2(prob)
        
    return ent
```

```python
def divide_data(data, attribute, threshold):
    left = pd.DataFrame([], columns = data.columns)
    right = pd.DataFrame([], columns = data.columns)
    
    for ix in range(data.shape[0]):
        if data[attribute].loc[ix] < threshold:
            left = left.append(data.loc[ix])
        else:
            right = right.append(data.loc[ix])
    
    return left, right

def info_gain_ent(data, ent, attribute, threshold):
    left, right = divide_data(data, attribute, threshold)
    
    ig = ent - entropy(left['profitable']) - entropy(left['profitable'])
    
    return ig
```
