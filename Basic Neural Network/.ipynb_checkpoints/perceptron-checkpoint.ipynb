```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, Y = make_blobs(n_samples=500, centers=2, n_features=2, random_state=10)
```

```python
print(X.shape, Y.shape)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def predict(X, W):
    z = np.dot(X, W)
    return sigmoid(z)
```

```python
def loss(X, Y, W):
    """binary cross entropy"""
    Y_ = predict(X, W)
    cost = np.mean(-Y*np.log(Y_) - (1-Y)*np.log(1-Y_))
    return cost
```

```python
def updateW(X, Y, W, learning_rate):
    Y_ = predict(X, W)
    grad = np.dot(X.T, Y_ - Y)
    
    m = X.shape[0]
    W = W - learning_rate*grad/float(m)
    
    return W
```

```python
def train(X, Y, learning_rate = 0.5, epochs = 500):
    ones = np.ones((X.shape[0], 1))
    X = np.hstack((ones, X))
    
    W  = np.zeros((X.shape[1])) ## n+1 entries
    
    for i in range(epochs):
        W = updateW(X, Y, W, learning_rate)
        
        if epochs%10 == 0:
            l = loss(X, Y, W)
            print("epoch %d loss is %.4f"%(i, l), end='\r')
    
    return W
```

```python
train(X, Y)
```
